```python
# This code completely wrote based on sudo code provided by priya bhatia.
# I changed only one condition in quick_sort method.
def partition(a, p, q):
    x = a[p]
    i = p
    for j in range(p + 1, q):
        if a[j] <= x:
            i = i + 1
            temp = a[i] # not need to use temp variable here, we can swap like this : a[i], a[j] = a[j], a[i]
            a[i] = a[j]
            a[j] = temp

    temp = a[p]
    a[p] = a[i]
    a[i] = temp
    # returning the i pointer (index)--> this is going to be value for "m" variable in quick_sort method.
    return i


def quick_sort(a, p, q):
    if a == p:
        return a[p]
    # if we don't use this(below) condition, the code will run continuously( bcz the if condition never satisfies until
    # the length of the array is 1.
    if p < q:
        m = partition(a, p, q)
        quick_sort(a, p, m)
        quick_sort(a, m + 1, q)
    # no need to use return statement here, the quick sort algo. follows (inplace sorting). So, it will automatically
    # change original array.
    return arr
```

```python
# Case : 3
 # Single number order
arr = [10 for i in range(10)]
quick_sort(arr, 0, len(arr))
print(arr)
```
